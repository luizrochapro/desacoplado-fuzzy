```python
from FuzzyInfSystem import *
import matplotlib.pyplot as plt

f = FuzzyInfSystem(100,2,0.1,0.01)

x = f.pert_funcs_df()
y = f.pert_funcs_dx()

```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(8, 9))

ax1.plot(f.uni_dis_F, x[0],  linewidth=1.5, label='LN')
ax1.plot(f.uni_dis_F, x[1],  linewidth=1.5, label='MN')
ax1.plot(f.uni_dis_F, x[2],  linewidth=1.5, label='SN')
ax1.plot(f.uni_dis_F, x[3],  linewidth=1.5, label='ZR')
ax1.plot(f.uni_dis_F, x[4],  linewidth=1.5, label='SP')
ax1.plot(f.uni_dis_F, x[5],  linewidth=1.5, label='MP')
ax1.plot(f.uni_dis_F, x[6],  linewidth=1.5, label='LP')
ax1.set_title('Funções de Pertinência DF')
ax1.legend()

ax2.plot(f.uni_dis_X, y[0],  linewidth=1.5, label='LN')
ax2.plot(f.uni_dis_X, y[1],  linewidth=1.5, label='MN')
ax2.plot(f.uni_dis_X, y[2],  linewidth=1.5, label='SN')
ax2.plot(f.uni_dis_X, y[3],  linewidth=1.5, label='ZR')
ax2.plot(f.uni_dis_X, y[4],  linewidth=1.5, label='SP')
ax2.plot(f.uni_dis_X, y[5],  linewidth=1.5, label='MP')
ax2.plot(f.uni_dis_X, y[6],  linewidth=1.5, label='LP')
ax2.set_title('Funções de Pertinência DX')
ax2.legend()

for ax in (ax1,ax2):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

```python
dp = fuzz.trimf(f.uni_dis_F, [0, 5, 10])

w = fuzz.fuzzy_and(f.uni_dis_F, dp, f.uni_dis_F, x[3])

plt.plot(f.uni_dis_F, w[1])
```
